```python
import os
import csv
import pprint
import pandas as pd
```

```python
file = "data/obj_pose-laser-radar-synthetic-input.txt"
radar_data = "radar.csv"
lidar_data = "lidar.csv"


data_r = open(radar_data, 'w')
data_l = open(lidar_data, 'w')


header_r = "sensor_type rho_measured phi_measured rhodot_measured timestamp x_groundtruth y_groundtruth vx_groundtruth vy_groundtruth yaw_groundtruth yawrate_groundtruth\n"
header_l = "sensor_type x_measured y_measured timestamp x_groundtruth y_groundtruth vx_groundtruth vy_groundtruth yaw_groundtruth yawrate_groundtruth\n"

data_r.write(header_r)
data_l.write(header_l)


f = open(file, 'r')

for line in f:
    if line.startswith("R"):
        data_r.write(line)
    else:
        data_l.write(line)
```

```python
with open(radar_data, 'r') as ff:
    for row in ff:
        pprint.pprint(row)
```
